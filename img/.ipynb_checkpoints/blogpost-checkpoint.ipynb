```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
t0 = 0.  # Initial time
tf = 15. # Final time
t_span = (t0, tf) # Time span

k0 = 2.5 # Spring specific strength
initial_position = 1. # Starting position x_0
initial_velocity = 0. # Starting velocity v_0

y0 = (initial_position, initial_velocity) # Initial values vector

# Generate times to sample the output solution at
time_steps = 10000
tcoordinates = np.linspace(t_span[0], t_span[1], time_steps) 
```

```python
def derivative_field(t, y, k0):
    return np.array([y[1], 
                     -k0*y[0]])
```

```python
# Vector field
X, V = np.meshgrid(np.arange(-1, 1, .1), np.arange(-1, 1, .1))
vX, vV = derivative_field(0., [X, V], k0)

magnitude = np.sqrt(vX**2 + vV**2)

fig1, ax1 = plt.subplots()
ax1.set_title('2D vector field of the harmonic ODE')
ax1.set(xlabel='Position $x$', ylabel='Velocity $y$')
Q = ax1.quiver(X, V, vX, vV, magnitude, cmap="viridis", scale=30, pivot='mid', width=0.006)

fig1.savefig("blog_h1_vecfield.svg")
plt.show()
```

```python
sol = sp.integrate.solve_ivp(derivative_field, t_span, y0, args=(k0,), 
                                    method='DOP853', rtol=1e-13, atol=1e-13, 
                                    max_step=1/100, t_eval=tcoordinates, 
                                    dense_output=True, events=None, vectorized=False)
```

```python
sol
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)

fig.suptitle('Harmonic Oscillator solution')

ax1.set(ylabel='Position $x$')
ax2.set(xlabel='time $t$', ylabel='Velocity $v$')

ax1.plot(tcoordinates, sol.y[0,:], "-", color="tab:blue")
ax2.plot(tcoordinates, sol.y[1,:], "-", color="tab:red")
```

```python
# Plot solution on
X, V = np.meshgrid(np.arange(-1.2, 1.2, .1), np.arange(-1.6, 1.6, .1))
vX, vV = derivative_field(0., [X, V], k0)

magnitude = np.sqrt(vX**2 + vV**2)

fig1, ax1 = plt.subplots()
ax1.set_title('2D vector field of the harmonic ODE')
ax1.set(xlabel='Position $x$', ylabel='Velocity $y$')
ax1.quiver(X, V, vX, vV, magnitude, cmap="viridis", scale=40, pivot='mid', width=0.003)

ax1.plot(sol.y[0,:], sol.y[1,:], "-", color="tab:blue")
ax1.scatter(sol.y[0,0], sol.y[1,0], color="tab:red")
vX0, vV0 = derivative_field(0., [sol.y[0,0], sol.y[1,0]], k0)
ax1.quiver(sol.y[0,0], sol.y[1,0], vX0, vV0, color="tab:red", scale=10, width=0.005)

fig1.savefig("blog_h1_vecfield_sol.svg")
plt.show()
```
